```
# >> G: SLICE THROUGH THE IMAGE and plot the raw and unsmoothed data
# 1. Examine the image you displayed earlier.  We'll copy a horizontal line
#    of data from 'tornado' image row 100 to use here.
# 2. Copy a row of data from tornado (row 100, all columns) and put this
#    line of data into array "raw"
# 3. View this web page:  https://stackoverflow.com/questions/5283649/plot-smooth-line-with-pyplot
#       ... and search for "ndimage".  Note the "from scipy.." line and way in which data is smoothed.
# 4. Using the above, pass your "raw" array to the gaussian_filter1d function, and
#    put the result in array "smooth"
# 5. Plot the original line of data ('raw'), and then plot on top of it'smooth'
#    Color these lines differently.  Based on this, try different "sigma=" settings
#    until the rough parts of the plot are gone (you will need a value > 2).
# 6. Add x-axis label Column and y-axis label Intensity.
# 7. Add a title that make sense (you choose)
# 8. Go back and use 'plt.figure' with figsize=(8,4) and rerun the cell.

raw=tornado[100,:]
from scipy.ndimage.filters import gaussian_filter1d

smooth = gaussian_filter1d(raw, sigma=4)
plt.plot(raw)
plt.plot(smooth)
plt.ylabel('Intensity')
plt.xlabel('Column')
plt.title('Tornado Intensity')
plt.figure(figsize=(8,4), dpi=80)
plt.show()
```

```
# >> H: PLOT FOUR IMAGES
# You'll plot images 5, 15, 25, and 35 here, using a loop.
# 1. Set variable "panel" to zero
# 2. Call this function to set the image dpi: plt.gcf().set_dpi(150)
# 3. LOOP (use 'for') with variable "nfile" running from 5 to 35 by 10
#   (but >including< 35, so your stopping number must be >35)
# 4. Inside the for loop, set variable 'filename' like this:
#      filename = 'LoneTree_Iowa2.000%02d.png' % nfile
#      (discussion: %02d makes zero-padded numbers 00, 01, ..., 09, 10, ...)
# 5. Read the image (whose name is: filename) into variable 'tornado'
# 6. Now make the value of variable 'panel' increase by 1.
# 7. Call plt.subplot, set up for a 2x2 figure, with 'panel' the third variable.
#     ... this will put the first plot at top left, second at top right, etc.
# 8. Call imshow to display 'tornado' again with a gray color map
# 9. Add a title that says "Image #" with "#" the file number (nfile)
# 10. Finish by calling tight_layout() outside of the for loop.

panel = 0
plt.gcf().set_dpi(150)
for nfile in range(5,36,10):
  filename = 'LoneTree_Iowa2.000%02d.png' % nfile
  tornado = mpimg.imread(filename)
  panel=panel+1
  plt.subplot(2,2,panel)
  plt.imshow(tornado, cmap='gray')
  plt.title('Image %02d'%nfile)
plt.tight_layout()
```
